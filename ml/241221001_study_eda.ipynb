## EDA 란?
EDA(Exploratory Data Analysis, 탐색적 데이터 분석) 는 데이터를 이해하고, 패턴을 발견하고, 가설을 설정하고, 통찰력을 얻기 위해 데이터를 분석하고 요약하는 중요한 단계입니다. EDA는 모델을 구축하기 전에 데이터를 시각화하고, 통계적 방법 등을 활용하여 데이터의 주요 특징을 파악하는 데 중점을 둡니다.

## EDA의 주요 목적

- 데이터 이해: 데이터의 구조, 변수 유형, 분포, 결측치 여부 등을 파악합니다.
- 패턴 발견: 변수 간의 관계, 이상치, 군집 등 숨겨진 패턴을 찾아냅니다.
- 가설 설정: 데이터 분석을 통해 얻은 통찰력을 바탕으로 모델링에 필요한 가설을 설정합니다.
- 모델 선택: 데이터의 특징을 이해하여 적합한 머신 러닝 모델을 선택하는 데 도움을 줍니다.
- 특징 추출 및 생성: 모델 성능을 향상시킬 수 있는 새로운 특징을 추출하거나 생성하는 데 도움을 줍니다.
- 데이터 전처리: 모델링에 앞서 필요한 데이터 전처리(결측치 처리, 이상치 제거, 변수 변환 등) 방향을 결정합니다.

## EDA에서 사용되는 주요 기법

- 기술 통계 (Descriptive Statistics):
  - 중심 경향 (Central Tendency): 평균(mean), 중앙값(median), 최빈값(mode) 등을 통해 데이터의 중심 경향을 파악합니다.
  - 변동성 (Variability): 분산(variance), 표준편차(standard deviation), 범위(range), 사분위수 범위(IQR) 등을 통해 데이터의 퍼짐 정도를 파악합니다.
  - 분포 형태 (Shape of Distribution): 왜도(skewness), 첨도(kurtosis) 등을 통해 데이터 분포의 비대칭성과 뾰족한 정도를 파악합니다.

- 데이터 시각화 (Data Visualization):
  - 히스토그램 (Histogram): 단일 변수의 분포를 시각적으로 나타냅니다.
  - 산점도 (Scatter Plot): 두 변수 간의 관계를 시각적으로 나타냅니다.
  - 박스 플롯 (Box Plot): 단일 변수의 분포를 요약하여 보여주며, 이상치를 쉽게 확인할 수 있습니다.
  - 히트맵 (Heatmap): 여러 변수 간의 상관 관계를 색상으로 나타내어 시각적으로 파악하기 쉽습니다.
  - 라인 그래프 (Line Graph): 시계열 데이터의 추세를 나타내는 데 유용합니다.

- 데이터 그룹화 (Data Grouping): 데이터를 특정 기준에 따라 그룹화하여 그룹 간의 차이를 비교하고 분석합니다.

- 차원 축소 (Dimensionality Reduction): 고차원 데이터를 저차원으로 축소하여 데이터의 주요 특징을 유지하면서 시각화 및 분석을 용이하게 합니다. (예: PCA, t-SNE)

```python
# 데이터의 상관계수 행렬을 생성
corr = data.corr(numeric_only=True) # 데이터셋으로부터 각 변수별 상관계수를 계산하여 행렬형태로 나타내는 함수. 여기서는 피어슨 상관계수를 이용
# figure에서 생략될 부분을 지정하는 mask 행렬을 생성
mask = np.ones_like(corr, dtype=bool)
mask = np.triu(mask)
# mask = np.tril(mask)

# 시각화될 그림의 크기를 지정
plt.figure(figsize=(13, 10))
# 히트맵 형태로 상관행렬 시각화하기
sns.heatmap(data=corr, annot=True, fmt='.2f', mask=mask, cmap='RdYlBu_r')
plt.show()
```

```python
# 분석을 완료한 데이터 샘플 123개의 인덱스 확인
indices_to_keep = data.isnull().sum(axis=1) < 8

# 샘플을 제거, 비교를 위해 원본 데이터는 별도로 복사
origin_data = data
data = data[indices_to_keep].copy()
print(f"샘플 데이터 수 변화: {len(origin_data)} -> {len(data)}")

# 나머지 결측치중 가장 많은 비중을 차지하는 subscribers_for_last_30_days에서 NaN을 0으로 대체
data['subscribers_for_last_30_days'].fillna(0, inplace=True)

# 나머지 결측치를 포함하는 모든 행들은 버림(drop)
data.dropna(inplace=True)
print(f"처리 완료 후 데이터 샘플 수: {len(data)}")

# 결측치가 잘 제거되었는지 확인
data.isnull().sum()
```

```python
# 데이터 분포 변환을 위한 scikit-learn 패키지 내 함수 불러오기
from sklearn.preprocessing import scale
```

```python
target_feature = 'lowest_monthly_earnings'
# 데이터 로그변환
# 로그 스케일로 변환한 값으로 새 변수를 만든다. 새 변수명은 기존 변수명에 prefix를 붙인다
data[f'log_{target_feature}'] = scale(np.log(data[target_feature] + 1)) # log에 0값이 들어가는 것을 피하기 위해 1을 더해준다
# 로그 변환한 변수의 요약통계량 보기
data[f'log_{target_feature}'].describe()
```

```python
# 로그 변환 전후의 데이터 분포 비교
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# 로그 변환 전 데이터 분포
sns.histplot(data=data, x=target_feature, kde=True, ax=axes[0])
axes[0].set_title(f'Before')

# 로그 변환 후 데이터 분포
sns.histplot(data=data, x=f'log_{target_feature}', kde=True, ax=axes[1])
axes[1].set_title('After')
# axes[1].set_xlabel(f'log_{target_feature}')
# axes[1].set_ylabel('빈도수')

# plt.tight_layout()
plt.show()
```

### 표준화

데이터의 분포를 평균이 0, 분산이 1이 되도록 변환하는 것

### 정규화

데이터를 최솟값이 0이고 최댓값이 1로, 0에서 1사이의 값을 가지도록 변환하는 방법

```python
# 데이터 분포 변환을 위한 scikit-learn 패키지 내 함수 불러오기
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
# 표준화를 사용해서 데이터의 단위를 변환
standard_scaler = StandardScaler()
data[f'standardized_{target_feature}'] = standard_scaler.fit_transform(data[[target_feature]])

# 비교를 위해 변환 전후의 피쳐를 선택
feature_original = data[target_feature]
feature_standardized = data[f'standardized_{target_feature}']

# 표준화된 변수의 요약통계량 보기
print(data[target_feature].describe())
print(data[f'standardized_{target_feature}'].describe())
```

```python
# 표준화된 데이터와 원본 데이터를 히스토그램으로 비교 시각화
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# 원본 데이터 히스토그램
sns.histplot(data=feature_original, kde=True, ax=axes[0])
axes[0].set_title(f'Original {target_feature}')

# 표준화된 데이터 히스토그램
sns.histplot(data=feature_standardized, kde=True, ax=axes[1])
axes[1].set_title(f'Standardized {target_feature}')

plt.tight_layout()
plt.show()
```

```python
# 정규화를 사용해서 데이터의 단위를 변환
min_max_scaler = MinMaxScaler()
data[f'normalized_{target_feature}'] = min_max_scaler.fit_transform(data[[target_feature]])

# 비교를 위해 변환 전후의 피쳐를 선택
feature_normalized = data[f'normalized_{target_feature}']

# 정규화된 변수의 요약통계량 보기
print(data[target_feature].describe())
print(data[f'normalized_{target_feature}'].describe())

# 정규화된 데이터와 원본 데이터를 히스토그램으로 비교 시각화
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# 원본 데이터 히스토그램
sns.histplot(data=feature_original, kde=True, ax=axes[0])
axes[0].set_title(f'Original {target_feature}')

# 정규화된 데이터 히스토그램
sns.histplot(data=feature_normalized, kde=True, ax=axes[1])
axes[1].set_title(f'Normalized {target_feature}')

plt.tight_layout()
plt.show()
```
